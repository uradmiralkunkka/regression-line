```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# 1. Buat data dummy
np.random.seed(42)
jumlah_data = 30
curah_hujan = np.random.uniform(50, 200, jumlah_data)  # mm
penjualan_kopi = 2 * curah_hujan + np.random.normal(0, 20, jumlah_data)  # kg

# Buat DataFrame
df = pd.DataFrame({
    'Curah Hujan': curah_hujan,
    'Penjualan Kopi': penjualan_kopi
})

# 2. Regresi Linier
X = df[['Curah Hujan']]
y = df['Penjualan Kopi']
model = LinearRegression()
model.fit(X, y)
df['Prediksi'] = model.predict(X)

# 3. Scatter plot dengan garis regresi
plt.figure(figsize=(10, 6))
plt.scatter(df['Curah Hujan'], df['Penjualan Kopi'], color='blue', label='Data Asli')
plt.plot(df['Curah Hujan'], df['Prediksi'], color='red', linewidth=2, label='Garis Regresi')
plt.xlabel('Curah Hujan (mm)')
plt.ylabel('Penjualan Kopi (kg)')
plt.title('Scatter Plot Penjualan Kopi vs Curah Hujan')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.savefig("garis regresi.png", dpi=300, bbox_inches='tight')  # Simpan gambar
plt.show()
```

```python
# Cetak persamaan regresi linear
slope = model.coef_[0]
intercept = model.intercept_
persamaan = f'y = {slope:.2f}x + {intercept:.2f}'
print("Persamaan linear antara curah hujan dan penjualan kopi:")
print(persamaan)
```
